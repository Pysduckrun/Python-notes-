```python
# new thing here is linestyle and linewidth

plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# markers and markersize
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

#plt.legend(loc='upper left')
plt.legend()
```

```python
# show -> we have to call it explicitly in certain text editors otherwise
# the graph won't be printed
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()

plt.show()
```

# plt.plot vs plt.scatter
### Key Breakdown
1. The "Connection" FactorThe fundamental difference is how they handle the relationship between data points.plt.plot() assumes your data has an inherent order (like time-series data or a mathematical function). It "draws a line" through your points to guide the eye.plt.scatter() assumes your data points are independent observations. It keeps them separate to show how they cluster or spread across the coordinate space.

2. Performance for Large DatasetsBecause plt.plot() connects points with simple line segments, it can handle hundreds of thousands of points efficiently. plt.scatter(), however, treats every point as a distinct object (with potentially unique sizes, colors, and edges), which consumes significantly more memory and processing power. If you are plotting 1,000,000 points, use plot or a specialized hexbin plot.

3. Individual Aestheticsplt.scatter() allows for "bubble charts." Because you can pass a list of sizes (s) and colors (c) equal to the length of your data, you can encode a third or fourth dimension of information into your plot. plt.plot() applies one style (e.g., "blue line with dots") to the entire data series.When to use which?Use plt.plot() when: You are plotting time-series data, scientific functions ($y = x^2$), or any data where the transition between points is as important as the points themselves.Use plt.scatter() when: You are performing exploratory data analysis, looking for correlations between variables, clustering, or handling multi-dimensional data where you want to represent an extra variable via marker size or color intensity.

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color=colors)
```

```python
# horizontal bar chart
plt.barh(colors,children,color=colors)
```

```python
# color and label
df = pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
# Multiple bar charts
# xticks
# arange
# width
plt.bar(np.arange(df.shape[0])- 0.2, df['2015'], width =0.2)
plt.bar(np.arange(df.shape[0]) , df['2016'], width =0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width =0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])
```

```python
np.arange(df.shape[0])
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation=90)
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2015'] + df['2016']),label='2017')
plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['english','hindi','maths','science','sst']
plt.pie(data,labels = subjects  )
plt.show()
```

```python
# dataset
df = pd.read_csv('/content/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'], autopct ='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0,0.5,0,0,0,0],shadow=True)
plt.show()
```
